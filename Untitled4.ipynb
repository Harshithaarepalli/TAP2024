```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
```

```python
# Load the dataset
data = pd.read_csv('fraudTest.csv')
```

```python
# Display the column names and the first few rows to identify the target variable
print(data.columns)
print(data.head())
```

```python
# Replace 'isFraud' with the correct target column name identified from the dataset inspection
target_column = 'is_fraud'  # Replace this with the correct column name
```

```python
# Drop rows with missing target values (if any)
data.dropna(subset=[target_column], inplace=True)
```

```python
# Separate numerical and categorical columns
numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = data.select_dtypes(include=[object]).columns.tolist()
```

```python
# Fill missing values in numerical features with the median
for col in numerical_features:
    data[col] = data[col].fillna(data[col].median())
```

```python
# Fill missing values in categorical features with the mode
for col in categorical_features:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
# Normalize numerical features
scaler = StandardScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
# Encode categorical variables
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

print(data.head())
```

```python
#EDA
# Plot the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x=target_column, data=data)
plt.title('Distribution of Fraudulent vs. Non-Fraudulent Transactions')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize distribution of key numerical features
key_features = numerical_features[:3]  # Adjust this list as needed
for feature in key_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Boxplot to identify outliers in transactionAmount
plt.figure(figsize=(6, 4))
sns.boxplot(x=target_column, y='amt', data=data)
plt.title('Boxplot of Transaction Amount by Fraud Status')
plt.show()
```

```python
# Create new features based on time
data['transaction_hour'] = pd.to_datetime(data['trans_date_trans_time']).dt.hour
data['transaction_day'] = pd.to_datetime(data['trans_date_trans_time']).dt.day
```

```python
# Drop the original transactionTime column
data.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
# Correlation analysis to select relevant features
correlation_matrix = data.corr()
correlation_target = abs(correlation_matrix[target_column])
relevant_features = correlation_target[correlation_target > 0.1].index.tolist()
```

```python
print(f'Relevant features: {relevant_features}')
```

```python
# Prepare the data
X = data[relevant_features].drop(columns=[target_column])
y = data[target_column]
```

```python
# Check the unique values and data type of the target variable
print(y.unique())
print(y.dtype)
```

```python
# Convert target variable to binary if necessary
# Adjust the lambda function based on actual unexpected values
y = y.apply(lambda x: 1 if x == 16 else 0)
```

```python
# Check the unique values after conversion
print(y.unique())
```

```python

# Ensure the target variable is categorical
y = y.astype(int)  # Or use y.astype('category') if the target variable should be a category
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Confirm the resampling
print("Resampled dataset shape:", X_train_resampled.shape, y_train_resampled.shape)
```

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}
```

```python
# Train and evaluate models
results = {}
for model_name, model in models.items():
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test)
    results[model_name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC AUC": roc_auc_score(y_test, y_pred)
    }
```
