```python
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='xgboost')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedKFold, GroupKFold, RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN
```

```python
# Load the dataset
file_path = 'fraudTest.csv'
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values before handling:")
print(missing_values[missing_values > 0])
```

```python
# Fill missing numerical values with the mean
numerical_cols = df.select_dtypes(include=np.number).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
# Fill missing categorical values with the mode
categorical_cols = df.select_dtypes(include='object').columns
fill_modes = {col: df[col].mode()[0] for col in categorical_cols}
df.fillna(fill_modes, inplace=True)
```

```python
# Check for missing values after handling
missing_values = df.isnull().sum()
print("Missing values after handling:")
print(missing_values[missing_values > 0])
```

```python
# Selecting numerical features for normalization
numerical_features = ['amt', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
# Encoding categorical features using LabelEncoder
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Display the preprocessed DataFrame
print("Preprocessed DataFrame:")
print(df.head())
```

```python
# Exploratory Data Analysis (EDA)
# 2.1 Visualizing the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='is_fraud', data=df)
plt.title('Distribution of Fraudulent vs Non-Fraudulent Transactions')
plt.xlabel('Is Fraud')
plt.ylabel('Count')
plt.show()
```

```python
# 2.2 Visualizing the distribution of transaction amounts
plt.figure(figsize=(8, 6))
sns.histplot(df['amt'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2.3 Correlation Matrix
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 2.4 Comparative Visualization: Fraudulent vs Non-Fraudulent Transactions by Category
# Assuming the 'category' column contains the original category names
original_categories = df['category'].unique()  # Save original category names
```

```python
# Encode the 'category' column
label_encoders['category'] = LabelEncoder()
df['category'] = label_encoders['category'].fit_transform(df['category'])
```

```python
# Create a mapping of the encoded values to original names
category_mapping = dict(zip(label_encoders['category'].transform(original_categories), original_categories))
```

```python
# Plot using the original category names
plt.figure(figsize=(12, 6))
sns.countplot(x='category', hue='is_fraud', data=df)
plt.title('Distribution of Categories for Fraudulent vs. Non-Fraudulent Transactions')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Is Fraud')
```

```python
# Replace x-tick labels with the original category names
plt.xticks(ticks=np.arange(len(category_mapping)), labels=[category_mapping[i] for i in range(len(category_mapping))], rotation=90, ha='right')
plt.show()
```

```python
# 2.5 Comparative Visualization: Transaction Amounts by Fraud Status
plt.figure(figsize=(12, 6))
sns.boxplot(x='is_fraud', y='amt', data=df)
plt.title('Transaction Amounts by Fraud Status')
plt.xlabel('Is Fraud')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
# Feature Engineering
# Creating new features based on the existing data
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['trans_hour'] = df['trans_date_trans_time'].dt.hour
df['trans_day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
```

```python
# Drop the original 'trans_date_trans_time' column
df.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
# Display the DataFrame with new features
print("DataFrame with new features:")
print(df.head())
```

```python
# Model Development
# Define features and target
X = df.drop(columns=['is_fraud'])
y = df['is_fraud']
```

```python
# Handling class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
# Define the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
```

```python
# Function to evaluate model performance
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=1)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    return accuracy, precision, recall, f1, roc_auc
```

```python
# Evaluate the XGBoost model
accuracy, precision, recall, f1, roc_auc = evaluate_model(xgb_model, X_train, X_test, y_train, y_test)
print(f'XGBoost - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}, ROC AUC Score: {roc_auc:.4f}')
```

```python
# Cross-Validation
# Using RepeatedKFold
rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=42)
scores = cross_val_score(xgb_model, X_res, y_res, cv=rkf, scoring='f1')
print(f'XGBoost - RepeatedKFold F1 Score: {np.mean(scores):.4f} ± {np.std(scores):.4f}')
```

```python
# Using GroupKFold
# Assuming 'cc_num' is the grouping feature, include it in the initial dataset and use it for GroupKFold
df['cc_num'] = df['cc_num'].astype(str)
groups = df['cc_num']  
gkf = GroupKFold(n_splits=5)
scores = cross_val_score(xgb_model, X, y, cv=gkf, groups=groups, scoring='f1')
print(f'XGBoost - GroupKFold F1 Score: {np.mean(scores):.4f} ± {np.std(scores):.4f}')
```

```python
# Handling class imbalance using ADASYN
adasyn = ADASYN(random_state=42)
X_res_adasyn, y_res_adasyn = adasyn.fit_resample(X, y)
```

```python
# Splitting the data into training and testing sets for ADASYN resampled data
X_train_adasyn, X_test_adasyn, y_train_adasyn, y_test_adasyn = train_test_split(X_res_adasyn, y_res_adasyn, test_size=0.2, random_state=42)
```

```python
# Evaluate the XGBoost model with ADASYN resampled data
accuracy, precision, recall, f1, roc_auc = evaluate_model(xgb_model, X_train_adasyn, X_test_adasyn, y_train_adasyn, y_test_adasyn)
print(f'XGBoost (ADASYN) - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}, ROC AUC Score: {roc_auc:.4f}')
```

```python
# Grid Search for Hyperparameter Tuning
param_dist = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300]
}
random_search = RandomizedSearchCV(estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss'), 
                                   param_distributions=param_dist, 
                                   n_iter=10, 
                                   cv=3, 
                                   scoring='f1', 
                                   random_state=42,
                                   n_jobs=-1)
random_search.fit(X_train, y_train)

print(f'Best parameters for XGBoost: {random_search.best_params_}')
```

```python
# Train the best model with the best parameters
best_model = random_search.best_estimator_
best_model.fit(X_train, y_train)
```

```python
# Predicting the target for the test set with the best model
y_pred_best = best_model.predict(X_test)
```

```python
# Evaluate model performance with the best model
accuracy_best = accuracy_score(y_test, y_pred_best)
precision_best = precision_score(y_test, y_pred_best, zero_division=1)
recall_best = recall_score(y_test, y_pred_best)
f1_best = f1_score(y_test, y_pred_best)
roc_auc_best = roc_auc_score(y_test, y_pred_best)

print(f'Best Model Accuracy: {accuracy_best:.4f}')
print(f'Best Model Precision: {precision_best:.4f}')
print(f'Best Model Recall: {recall_best:.4f}')
print(f'Best Model F1 Score: {f1_best:.4f}')
print(f'Best Model ROC AUC Score: {roc_auc_best:.4f}')
```
