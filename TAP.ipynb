```python
import seaborn as sns
```

```python
fraud_job_title_df = fraud_job_title.reset_index()
fraud_job_title_df.columns = ['job', 'count']

# Plotting
plt.figure(figsize=(6, 4))
sns.barplot(data=fraud_job_title_df, x="count", y="job",palette="plasma")
plt.title('Top 10 Job Titles Associated with Fraud')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.xticks()
plt.show()
```

```python
df["category"].value_counts()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(x = "category",hue = "is_fraud",data = df,palette = "viridis")
plt.title("Category")
plt.xticks(rotation = 60)
plt.show()
```

```python
plt.figure(figsize = (8,6))
fraud_category = df[df["is_fraud"] == 1]["category"].value_counts()
sns.barplot(x = fraud_category.index,y = fraud_category.values, palette = "viridis")
plt.title("Category vs Fraud")
plt.xticks(rotation = 60)
plt.show()
```

```python
df["city"].value_counts()
```

```python
amt_99 = df['amt'].quantile(0.99)
plt.figure(figsize=(8,6))
sns.histplot(data=df[df['amt'] <= amt_99], x="amt", hue="is_fraud", multiple="stack", bins=50, palette="viridis")
plt.title('Distribution of Transaction Amounts (Up to 99th Percentile)')
plt.xlabel('Transaction Amount')
plt.ylabel('Count')
plt.show()
```

```python
fraud_rate_by_state = df.groupby('state')['is_fraud'].mean().sort_values(ascending=False).reset_index().head(10)

plt.figure(figsize=(8,6))
sns.barplot(data=fraud_rate_by_state, x='is_fraud', y='state', palette='viridis')
plt.title('Fraud Rate by State')
plt.xlabel('Fraud Rate')
plt.ylabel('State');
```

```python
df.set_index('trans_date_trans_time', inplace=True)
```

```python
plt.figure(figsize=(15, 8))
df.resample('D')['is_fraud'].sum().plot(label='Fraudulent Transactions', color='red')
df.resample('D')['is_fraud'].count().plot(label='Total Transactions', color='blue')
plt.title('Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df[df['is_fraud'] == 1], x='long', y='lat', hue='is_fraud', palette='viridis', legend=False)
plt.title('Fraudulent Transaction Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
